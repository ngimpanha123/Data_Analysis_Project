```python
from urllib.request import Request, urlopen, quote
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
class Car_Khmer24_Scrapping:
    
    from urllib.request import Request, urlopen, quote
    from bs4 import BeautifulSoup as bs
    
    def __init__(self, pages):
        self.pages = pages
        
    def get_html(self, url, user_agent = False):
        if user_agent: 
            user_agent = user_agent
        else : user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"

        request = Request(url, headers= {'User-Agent': user_agent})
        get_open = urlopen(request).read()
#         get_open = requests.get(url, headers = {'User-Agent': user_agent})
        
        return bs(get_open)
    
    def get_list_url_car(self):       
        link = []
        pages = self.pages
        for page in range(0,pages*50, 50):
            
            url = f"https://www.khmer24.com/en/motorcycles/honda/other.html?per_page={page}"
            html = self.get_html(url)
            list_item = html.find_all('li', class_="item")

    
            for i in range(len(list_item)):
                try:
                    car_url = quote(list_item[i].a.get('href'), safe='/:')
                    link.append(car_url)
                except: pass
        return link
    
    def get_car_data(self):
        
        links = self.get_list_url_car()
        description = {}
        for link in links:
            try:
                html = self.get_html(link)
                item_short_description = html.find('div', class_="item-short-description p-3 position-relative")
                item_short_description2 = html.find('ul', class_="list-unstyled item-fields")

                title_car = item_short_description.h1.get_text()

                price = html.find('div', class_="item-short-description p-3 position-relative").div.p.get_text()

                list_description = item_short_description.find_all('li')+item_short_description2.find_all('li')

                title_list = []
                value_list = []
                for i in range(len(list_description)):
                    title = list_description[i].find('span', class_ = 'title').get_text()
                    value = list_description[i].find('span', class_ = 'value').get_text()

                    if ':' in title: title = title.replace(':', '')

                    title_list.append(title)
                    value_list.append(value)


                    if not description or title not in description: 
                        description[title] = [value]
                    else: description[title].append(value)

                keys = description.keys()       

                key_list = ['Ad ID ', 'Category ', 'Locations ', 'Posted ', 'Bedroom', 'Bathroom', 'Size(m2)', 'Link', 'Title', 'Price', 'Type', 'Facing']
                for key in key_list:
                    if key not in list(keys):
                        description[key] = []

                for key in keys:
                    if key not in title_list+['Link', 'Title', 'Price']:
                        description[key].append('NaN')
                values = [link,title_car, price]       
                for i, title in enumerate(['Link', 'Title', 'Price']):
                    
                    if not description or title not in description:
                        description[title] = [values[i]]
                    else: description[title].append(values[i])
                    
            except: pass
        

        return description
        
        
```

```python
def get_car_data(links):
    description = {}
    prices, title_cars = [], []
    for link in links:
        try:
            htm = get_html(link)
            item_short_description = htm.find('div', class_="item-short-description p-3 position-relative")
            item_short_description2 = htm.find('ul', class_="list-unstyled item-fields")

            title_car = item_short_description.h1.get_text()
            title_cars.append(title_car)

            price = htm.find('div', class_="item-short-description p-3 position-relative").div.p.get_text()
            prices.append(price)

            list_description = item_short_description.find_all('li')+item_short_description2.find_all('li')

            title_list = []
            value_list = []
            for i in range(len(list_description)):
                title = list_description[i].find('span', class_ = 'title').get_text()
                value = list_description[i].find('span', class_ = 'value').get_text()

                if ':' in title: title = title.replace(':', '')

                title_list.append(title)
                value_list.append(value)


                if not description or title not in description: 
                    description[title] = [value]
                else: description[title].append(value)
     
            keys = description.keys()       

            key_list = ['Ad ID ', 'Category ', 'Locations ', 'Posted ', 'Car Makes', 'Car Model', 'Year', 'Tax Type', 'Condition', 'Fuel', 'Transmission', 'Color', 'Body Type']
            for key in key_list:
                if key not in list(keys):
                    description[key] = []

            for key in keys:
                if key not in title_list:
                    description[key].append('NaN')
        except: pass                      
    for i, title in enumerate(['Link', 'Title', 'Price']):
        value = [links,title_cars, prices]
        for j in range(len(value[i])):
            if not description or title not in description:
                description[title] = [value[i][j]]
            else: description[title].append(value[i][j])
    return description
```
